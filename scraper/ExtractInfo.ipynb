```python
import csv
import time
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup


outfile = open('productinfo','w')
writer = csv.writer(outfile)
writer.writerow(["ProductID","ProductName"])


count = 0
with open('productid.csv') as infile:
    for row in infile:
        
        #if count >= 10:
        #    break
        
        productid = row
        url="https://www.amazon.com/dp/" + productid
        req = Request(url,headers={'User-Agent' : "Magic Browser"})
        webpage = urlopen(req).read()
        soup=BeautifulSoup(webpage,'html.parser')
        info=soup.findAll("span", { "id" : "productTitle" })
        title = info[0].text
        
        writer.writerow([row[0:9], title])
        count += 1
        #print(title)
        print ("Finished " + str(count) + ", total process: " + str("%.5f"%(count/63002*100))+"%")
        if count % 100 == 0:
            print('Reach 100 products, rest 120 sec ')
            time.sleep(120) 
        
print(count)
```
